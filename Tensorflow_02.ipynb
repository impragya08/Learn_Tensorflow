# Indexing Tensors


*Tensors can be indexed just like Python lists.*

```python
#Getting the first 2 elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
#Get the first element from each dimension from each index except for the final one
rank_4_tensor[:1,:1,:1]    #[:1,:1,:1,:]
#except for the second last one
rank_4_tensor[:1,:1,:,:1]
```

```python
#Create a rank 2 tensor (2 dimensions)
rank_2_tensor=tf.constant([[10,3],
                           [3,4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
#Get the last element of each row of our rank 2 tensor
rank_2_tensor[:,-1]
```

```python
#Add in extra dimension to our rank_2_tensor
#... indicates every axis before the last one
#we can also use : : :,tf.newaxis
rank_3_tensor= rank_2_tensor[..., tf.newaxis]       #add an additional axis
rank_3_tensor
```

```python
#Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1)   #"-1" means expand the final axis
tf.expand_dims(rank_2_tensor, axis=0)   #expand the 0-axis
```

# 2-Matrix Multiplication

```python
#Matrix multiplication in tensorflow
tf.matmul(tensor,tensor)
```

```python
#Matrix multiplication with python operator "@"
tensor @ tensor
```

```python
#Matrix multiplication on same shapes
#Creating a tensor (3,2) tensor
X=tf.constant([[1,2],
               [3,4],
               [5,6]])
#Creating a tensor (3,2)tensor
Y=tf.constant([[8,9],
               [10,11],
               [12,13]])

#Matrix multiply of same shape
#INCOMPATIBLE
```

```python
#Correcting Code
#Lets change shape of Y
tf.reshape(Y, shape=(2,3))
```

```python
#Try to multiply X with reshaped Y
X @ tf.reshape(Y, shape=(2,3))
#OR
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
#If we reshape X
tf.matmul(tf.reshape(X, shape=(2,3)),Y)
```

```python
# Can do the same with transpose
#Transpose fix the axis whereas reshapes shuffle the numbers around the mentioned shapes
tf.transpose(X), tf.reshape(X, shape=(2,3))
```

# 3- The Dot Product
 *Matrix multiplication is also referred to as dot product.*

 
You can perform matrix multiplication using:
- tf.matmul()
- tf.tensordot()

```python
#Perform dot product on X and Y
#requires X and Y transposes
tf.tensordot(tf.transpose(X),Y, axes=1)
```

```python
#Perform matrix multiplication between X and Y(transposed)
tf.tensordot(X,tf.transpose(Y), axes=1)
```

# Changing Data Types of Tensors

```python
#Create a new tensor with default data type
B= tf.constant([1.7,1.8])
B.dtype
```

```python
#Change from float32 to float16
#Reduced Precision
B = tf.cast(B, dtype=tf.float16)
B.dtype
```

```python
#Change from int32 to float 32
C=tf.constant([7,10])
E= tf.cast(C, dtype=tf.float32)
E.dtype
```
